```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests
```

```python
np.random.seed(613)
x = np.arange(50)
y = np.random.randint(0, 100, 50)
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x, y)
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.bar(x, y)
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x, y)
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.hist(y)
```

```python
ax.set_title('Total growth over time')
ax.set_xlabel('Years since start')
ax.set_ylabel('Total growth')
fig.tight_layout()
fig
```

```python
font1 = {'family':'sans-serif','color':'blue','size':20}
font2 = {'family':'monospace','color':'green','size':14}
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y)
ax.set_title('Total growth over time', fontdict = font1)
ax.set_ylabel('Total growth', fontdict = font2)
ax.set_xlabel('Years since start', fontdict = font2)
fig.tight_layout()
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y)
ax.set_ylabel('Total growth', fontdict = font2)
ax.set_xlabel('Years since start', fontdict = font2)
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x, 
           y, 
           marker='*', # NEW ADDITIONS 
           color = "indigo") 
           # NEW ADDITIONS
fig.show()
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x, y, marker='*', color = 'indigo', linestyle = '--', # NEW ADDITIONS 
        linewidth = 2) 
        # NEW ADDITIONS
fig.show()
```

```python
fig, ax = plt.subplots(figsize=(5, 3))# color being the new addition

ax.plot(x,
        y,
        marker='*', 
        color = '#7425b9', 
        linestyle = '--', 
        linewidth = 2)
fig.show()
```

```python
ax.plot(x,y,marker='*', 
        markersize = 12, # NEW ADDITIONS 
        color = '#7425b9', 
        linestyle = '--', 
        linewidth = 2, 
        markeredgecolor = '#fa9359', # NEW ADDITIONS 
        markerfacecolor = '#000000' ) # NEW ADDITIONS
```

```python
ax.grid(axis = 'y')
ax.grid(axis = 'y', color = "blue", linewidth = 2, linestyle = '-.')
```
