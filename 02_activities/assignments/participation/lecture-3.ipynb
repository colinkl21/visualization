```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests
np.random.seed(613)
x = np.arange(50)
y1 = np.random.randint(0, 100,50)
y2 = np.random.randint(0, 100,50)
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y1)
ax.plot(x,y2)
fig.show()
6
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y1, label = "Person 1" )
ax.plot(x,y2, label = "Person 2" )
ax.legend(loc='lower right', frameon = True, #add frame around the legend 
          fontsize = 12, #change font size 
          ncol = 2, #specify number of columns 
          shadow = True )
fig.show()
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y1, label = "Person 1")
ax.plot(x,y2, label = "Person 2")
ax.legend(loc='upper left', bbox_to_anchor=(1, 1))
fig.show()
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')
#ax.text(10, 95, "This value is important!")
#fig.show()

ax.text(10, 95, "This value is important!", ha='center', va='center', color='red', size=20)
fig.show()
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')
#ax.annotate('This is important!', xy=(10, 95), xytext=(20, 94), arrowprops=dict(facecolor='black'))
#fig.show()
ax.annotate(
    'This is important!',
    xy=(10, 95), xytext=(20, 94),
    arrowprops=dict(arrowstyle="wedge", color="hotpink")
)
fig.show()
```
