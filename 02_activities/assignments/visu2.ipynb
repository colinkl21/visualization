```python


import pandas as pd
import matplotlib.pyplot as plt


df = "c:/Users/Colin/Downloads/toronto.crime.csv"   # change to your actual file path (CSV or Excel)

df = pd.read_csv(df)

df.describe()
```

```python
# keep only rows where SEXUAL_ORIENTATION_BIAS is not "None"

filtered = df[df["SEXUAL_ORIENTATION_BIAS"].notna()]

filtered.describe()
```

```python

offence_counts = filtered["PRIMARY_OFFENCE"].value_counts()


fig, ax = plt.subplots(figsize=(12, 8))

offence_counts.plot(kind="bar", ax=ax, color="#79305a", edgecolor="black")

ax.set_title("Sexual Orientation Motivated Offence", fontsize=20, fontweight="bold")
ax.set_xlabel("Offence Type", fontsize=12)
ax.set_ylabel("Count", fontsize=12)
plt.xticks(rotation=60, ha="right")

plt.show()
```
